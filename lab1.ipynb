```python
# Import the Image class from the IPython.display module
from IPython.display import Image

# Import the Image class from the PIL (Python Imaging Library) module
from PIL import Image

# Define the path to the image file
image_path = './SBsir.jpg'

# Open the image file using the Image class from PIL
original_image = Image.open(image_path)

# Resize the opened image to 256x256 pixels
resized_image = original_image.resize((256, 256))

# Display the resized image
resized_image.show()

# Save the resized image
display(resized_image)
```

```python
# Import the numpy and matplotlib.pyplot libraries
import numpy as np
import matplotlib.pyplot as plt

# Convert the resized image to a numpy array
image_array = np.array(resized_image)

# Split the image array into its color channels
red_channel = image_array[:, :, 0]
green_channel = image_array[:,:, 1]
blue_channel = image_array[:,:, 2]

# Create a figure with 3 subplots, one for each color channel
fig, axs = plt.subplots(1, 3, figsize=(10, 5))

# Display the red channel in the first subplot
axs[0].imshow(red_channel, cmap='Reds')
axs[0].set_title('Red Channel')
axs[0].axis('off')

# Display the green channel in the second subplot
axs[1].imshow(green_channel, cmap = 'Greens')
axs[1].set_title('Green Channel')
axs[1].axis('off')

# Display the blue channel in the third subplot
axs [2].imshow(blue_channel, cmap='Blues')
axs[2].set_title('Blue Channel')
axs[2].axis('off')

# Display the figure
plt.show()
```

```python
# Convert the resized image to grayscale
grey_scale = resized_image.convert('L')

# Display the grayscale image
display(grey_scale)
```

```python
# Import the OpenCV library
import cv2

# Convert the resized image to a NumPy array
image_np = np.array(resized_image)

# Convert the image from RGB to HSV using OpenCV
hsv_img = cv2.cvtColor(image_np, cv2.COLOR_RGB2HSV)

# Display the resized HSV image
plt.imshow(hsv_img)
plt.axis('off') # Hide axis numbers and ticks
plt.show()
```
